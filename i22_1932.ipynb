```python
# Question 2

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("earth_surface_temperatures.csv")
df
```

```python
df.isnull()
df['Temperature'] = df['Temperature'].fillna(df['Temperature'].mode())
df['Monthly_variation'] = df['Monthly_variation'].fillna(df['Monthly_variation'].mode())
df['Anomaly'] = df['Anomaly'].fillna(df['Anomaly'].mode())
df['Anomaly']
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Month'].astype(str) + '-' + df['Years'].astype(str), format='%m-%Y')
df
```
